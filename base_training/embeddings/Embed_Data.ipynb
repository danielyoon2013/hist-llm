<h1> Simple Embeddings </h1>

```python
import pandas as pd
from pathlib import Path
from sentence_transformers import SentenceTransformer
from tqdm.auto import tqdm

# --- CONFIG ---
INPUT_DIR = Path(r"C:\Users\danielyoon\Dropbox\hist_LLM\Data\Label_Data")
OUTPUT_DIR = Path(r"C:\Users\danielyoon\Dropbox\hist_LLM\Data\Embedding_Data")
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

# Small, efficient model (384-dimensional vectors)
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
def create_embeddings():
    # Find all labeled parquet files
    files = list(INPUT_DIR.glob("labeled_data_*.parquet"))
    
    for file_path in files:
        print(f"Processing: {file_path.name}")
        df = pd.read_parquet(file_path)
        
        # Ensure 'score' is treated as our target 'labels'
        if 'score' in df.columns:
            df = df.rename(columns={'score': 'labels'})
        
        # 1. Generate Embeddings (The X Variables)
        # batch_size 32 is safe for local CPU/RAM
        embeddings = model.encode(df['text'].tolist(), batch_size=32, show_progress_bar=True)
        
        # 2. Store as a list of arrays in the dataframe
        df['embedding'] = list(embeddings)
        
        # 3. Save only necessary columns for 3.b to save space
        # We keep original_index for traceability
        output_df = df[['original_index', 'labels', 'embedding']]
        
        output_filename = file_path.name.replace("labeled_data_", "embeddings_")
        output_df.to_parquet(OUTPUT_DIR / output_filename)
        print(f"Saved to: {output_filename}")
```

```python
if __name__ == "__main__":
    create_embeddings()
```

<h1> Longer Embeddings </h1>

```python
import pandas as pd
import torch
from pathlib import Path
from sentence_transformers import SentenceTransformer
from tqdm.auto import tqdm
```

```python
# --- CONFIG ---
INPUT_DIR = Path(r"C:\Users\danielyoon\Dropbox\hist_LLM\Data\Label_Data")
OUTPUT_DIR = Path(r"C:\Users\danielyoon\Dropbox\hist_LLM\Data\Embedding_Data")
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

# Determine device (Use GPU if available, else CPU)
device = "cuda" if torch.cuda.is_available() else "cpu"

# Load model
model = SentenceTransformer('BAAI/bge-large-en-v1.5', device=device)
model.max_seq_length = 512 # BGE's standard limit
```

```python
def create_embeddings():
    files = list(INPUT_DIR.glob("labeled_data_*.parquet"))
    
    for file_path in files:
        print(f"\nProcessing: {file_path.name} on {device}")
        df = pd.read_parquet(file_path)
        
        if 'score' in df.columns:
            df = df.rename(columns={'score': 'labels'})
        
        # --- BGE SPECIFIC STEP ---
        # BGE models perform better if you add this instruction to the text
        instruction = "Represent this historical text for quality classification: "
        texts = [instruction + str(t) for t in df['text'].tolist()]
        
        # 1. Generate Embeddings
        # We use convert_to_numpy=True to ensure compatibility with Lasso/Ridge later
        embeddings = model.encode(
            texts, 
            batch_size=16, # Reduced from 32 for 'large' model stability
            show_progress_bar=True,
            convert_to_numpy=True
        )
        
        # 2. Store as a list of arrays
        df['embedding'] = list(embeddings)
        
        # 3. Save
        output_df = df[['original_index', 'labels', 'embedding']]
        output_filename = file_path.name.replace("labeled_data_", "embeddings_bge_")
        
        output_df.to_parquet(OUTPUT_DIR / output_filename)
        print(f"Saved: {output_filename} (Dimensions: {embeddings.shape[1]})")
```

```python
if __name__ == "__main__":
    create_embeddings()
```

```python
def get_openai_embeddings(text_list, batch_size=25):
    all_embeddings = []
    # Use the encoding for text-embedding-3 models
    encoding = tiktoken.get_encoding("cl100k_base")
    
    # 1. Truncate each text locally BEFORE sending to API
    safe_text_list = []
    for text in text_list:
        tokens = encoding.encode(text)
        if len(tokens) > 7500:
            # Keep only the first 8000 tokens
            text = encoding.decode(tokens[:7500])
        safe_text_list.append(text.replace("\n", " "))

    # 2. Process in batches as before
    for i in tqdm(range(0, len(safe_text_list), batch_size), desc="Encoding Batches"):
        batch = safe_text_list[i : i + batch_size]
        response = client.embeddings.create(input=batch, model=MODEL_NAME)
        all_embeddings.extend([data.embedding for data in response.data])
        
    return all_embeddings

def create_embeddings():
    files = list(INPUT_DIR.glob("labeled_data_*.parquet"))
    # files = files[:1]
    files = [
        WindowsPath('C:/Users/danielyoon/Dropbox/hist_LLM/Data/Label_Data/labeled_data_1701_1725.parquet'),
        WindowsPath('C:/Users/danielyoon/Dropbox/hist_LLM/Data/Label_Data/labeled_data_1976_2000.parquet'),
        WindowsPath('C:/Users/danielyoon/Dropbox/hist_LLM/Data/Label_Data/labeled_data_2001_2023.parquet')
    ]
    
    for file_path in files:
        print(f"\nProcessing: {file_path.name}")
        df = pd.read_parquet(file_path)
        
        # Rename target column for Step 3.b
        if 'score' in df.columns:
            df = df.rename(columns={'score': 'labels'})
        
        # 1. Generate Embeddings via API
        print(f"Requesting embeddings for {len(df)} rows...")
        embeddings = get_openai_embeddings(df['text'].tolist(), batch_size=5)
        
        # 2. Store in the dataframe
        df['embedding'] = embeddings
        
        # 3. Save for Lasso/XGBoost consumption
        output_df = df[['original_index', 'labels', 'embedding']]
        output_filename = file_path.name.replace("labeled_data_", "embeddings_openai_")
        
        output_df.to_parquet(OUTPUT_DIR / output_filename)
        print(f"Successfully saved to: {output_filename}")
```
